```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import csv
from scipy import stats
```

```python
path = "./Chicago_Crimedata.csv"
chicagocrime_df = pd.read_csv(path)
chicagocrime_df.head()
```

```python
chicagocrime_df = pd.DataFrame({
    "ID":chicagocrime_df["ID"],
    "Year":chicagocrime_df["Year"],
    "Primary Type":chicagocrime_df["Primary Type"],
    "Domestic":chicagocrime_df["Domestic"],
    "Latitude":chicagocrime_df["Latitude"],
    "Longitude":chicagocrime_df["Longitude"],
    "Location":chicagocrime_df["Location"]
})
chicagocrime_df.head()
```

```python
chicagocrime_df = chicagocrime_df.dropna(how='any')
chicagocrime_df.head()
```

```python
chicagocrime_df.count()
```

```python
primary_group = chicagocrime_df.groupby("Primary Type")

crime_count = primary_group ["Year"].count().sort_values(ascending = False)
crime_count.head()

topfive = crime_count.head()
topfive
```

```python
by_year = chicagocrime_df.groupby(["Year"])["Primary Type"].count()
by_year 
```

```python
#THEFT
theft_table = chicagocrime_df.loc[chicagocrime_df["Primary Type"]=="THEFT"]
theft_count = theft_table.groupby(["Year"])["Primary Type"].count()

#BATTERY
battery_table = chicagocrime_df.loc[chicagocrime_df["Primary Type"]=="BATTERY"]
battery_count = battery_table.groupby(["Year"])["Primary Type"].count()

# NARCOTICS
narcotics_table = chicagocrime_df.loc[chicagocrime_df["Primary Type"]=="NARCOTICS"]
narcotics_count = narcotics_table.groupby(["Year"])["Primary Type"].count()

# CRIMINAL DAMAGE 
criminal_damage_table = chicagocrime_df.loc[chicagocrime_df["Primary Type"]=="CRIMINAL DAMAGE"]
criminal_damage_count = criminal_damage_table.groupby(["Year"])["Primary Type"].count()
criminal_damage_count

# DECEPTIVE PRACTICE 
deceptive_practice_table = chicagocrime_df.loc[chicagocrime_df["Primary Type"]=="DECEPTIVE PRACTICE"]
deceptive_practice_count = deceptive_practice_table.groupby(["Year"])["Primary Type"].count()

crime_df = pd.DataFrame({"THEFT":theft_count,
                        "BATTERY":battery_count,
                        "NARCOTICS":narcotics_count,
                        "CRIMINAL DAMAGE":criminal_damage_count,
                        "DECEPTIVE PRACTICE":deceptive_practice_count})
crime_df.head()
```

```python
crime_df['Years'] = crime_df.index
crime_df.index
crime_df
```

```python
#drop 2019 data
crime_df  = crime_df [crime_df ['Years'] != 2019]
crime_df
```

```python
year = crime_df.iloc[:,5]
```

```python
#theft crime rate

theft_crime_rate = crime_df.iloc[:, 0]
tc_slope, tc_int, tc_r, tc_p, tc_std_err = stats.linregress(
   year, theft_crime_rate)
tc_fit = tc_slope * year + tc_int

#battery crime rate

battery_crime_rate = crime_df.iloc[:, 1]
bc_slope, bc_int, bc_r, bc_p, bc_std_err = stats.linregress(
   year, battery_crime_rate)
bc_fit = bc_slope * year + bc_int

#nacortics crime rate

nacortics_crime_rate = crime_df.iloc[:, 2]
nc_slope, nc_int, nc_r, nc_p, nc_std_err = stats.linregress(
   year, nacortics_crime_rate)
nc_fit = nc_slope * year + nc_int

#criminal damage crime rate

criminal_damage_crime_rate = crime_df.iloc[:, 3]
cdc_slope, cdc_int, cdc_r, cdc_p, cdc_std_err = stats.linregress(
   year, criminal_damage_crime_rate)
cdc_fit = cdc_slope * year + cdc_int

#deceptive practice crime rate

deceptive_practice_rate = crime_df.iloc[:, 4]
dpc_slope, dpc_int, dpc_r, dpc_p, dpc_std_err = stats.linregress(
   year, deceptive_practice_rate)
dpc_fit = dpc_slope * year + dpc_int
```

```python
# Plot
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,figsize=(15,12), sharex=True)
fig.suptitle("Crime Rates Over Time", fontsize=16, fontweight="bold" , )




ax1.set_xlim(min(year), max(year))
ax1.plot(year, theft_crime_rate, linewidth=1, color="r", marker="o")
ax1.plot(year, tc_fit, "b--", linewidth=1)
ax1.set_ylabel("Theft Crime Rate")

ax2.plot(year, battery_crime_rate, linewidth=1, marker="o", color="r")
ax2.plot(year, bc_fit, "r--", linewidth=1)
ax2.set_ylabel("Battery Crime Rate")

ax3.plot(year, nacortics_crime_rate, linewidth=1, marker="o", color="g")
ax3.plot(year, nc_fit, "g--", linewidth=1)
ax3.set_ylabel("Narcotics Crime Rate")


ax4.plot(year, criminal_damage_crime_rate, linewidth=1, marker="o", color="c")
ax4.plot(year, cdc_fit, "c--", linewidth=1)
ax4.set_ylabel("Criminal Damage Rate")

ax5.plot(year, deceptive_practice_rate, linewidth=1, marker="o", color="y")
ax5.plot(year, dpc_fit, "y--", linewidth=3)
ax5.set_ylabel("Deceptive Practice Rate")
ax5.set_xlabel("Year")
```
