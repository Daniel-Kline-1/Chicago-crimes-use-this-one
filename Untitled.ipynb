```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import csv
```

```python
path = "Chicago_Crimedata.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
dff = pd.DataFrame({
    "ID":df["ID"],
    "Year":df["Year"],
    "Primary Type":df["Primary Type"],
    "Latitude":df["Latitude"],
    "Longitude":df["Longitude"],
    "Location":df["Location"],
    "Location Description":df["Location Description"]
})
dff = dff.dropna(how ="any")
dff.head()
```

```python
len(dff)
```

```python
years_data = []
for year in list(range(2001,2015)):
    years_data.append(dff.loc[dff["Year"]== year,:])
```

```python
years_data[0].head()
```

```python
type_data = []
type_list = list(dff["Primary Type"].unique())

for types in type_list:
    type_data.append(dff.loc[dff["Primary Type"] == types,:])
```
