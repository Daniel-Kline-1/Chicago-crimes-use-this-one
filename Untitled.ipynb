```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import csv
```

```python
path = "Chicago_Crimedata.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
dff = pd.DataFrame({
    "ID":df["ID"],
    "Year":df["Year"],
    "Primary Type":df["Primary Type"],
    "Location Description":df["Location Description"],
    "Latitude":df["Latitude"],
    "Longitude":df["Longitude"],
    "Location":df["Location"]
})
dff.head()
dff.count()
```

```python
years_data = []
for year in list(range(2001,2018)):
    years_data.append(df.loc[dff["Year"]== year,:])
```

```python
years_data[0]
```

```python
# drop null rows
no_null_crime_df = dff.dropna(how='any')
no_null_crime_df
no_null_crime_df.count()
```

```python
no_null_crime_df['Primary Type'].value_counts()
```

```python
crime_by_year_df = no_null_crime_df['Year'].value_counts(ascending=True)
fig=plt.figure(figsize=(10, 8))
plt.title("Crimes Committed By Year")
plt.xlabel("Frequency of Crimes")
plt.ylabel("Year")
ax = crime_by_year_df.plot(kind='barh')
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
```

```python
# ***Use dataset top determine effectiveness of past Chicago PD actions to lower crime***
#Show how successful arreststs have been in Chicago over the years (by percentage). 
```

```python
arrests_by_year = df.groupby('Year')['Arrest'].value_counts().rename('Counts').to_frame()
arrests_by_year['Percentage'] = (100 * arrests_by_year / arrests_by_year.groupby(level=0).sum())
arrests_by_year.reset_index(level=[1],inplace=True)
arrests_by_year
```

```python
#Changes in crime in Chicago over the years
lineplt_arrests_years = arrests_by_year[arrests_by_year['Arrest'] == True]['Percentage']

labels = lineplt_arrests_years.index.values
fig=plt.figure(figsize=(12, 10))
plt.title('Successful Arrests By Year')
plt.xlabel("Year")
plt.ylabel("Percentage of Successful Arrests")
plt.xticks(lineplt_arrests_years.index, lineplt_arrests_years.index.values)

lineplt_arrests_years.plot(grid=True, marker='o', color='mediumvioletred')
```

```python
#Use District to group crime by area
```

```python
crime_by_district = df['District'].value_counts(ascending=True)
crime_by_district.index = crime_by_district.index.astype(int)
crime_by_district.index = crime_by_district.index.astype(str)
```

```python
fig=plt.figure(figsize=(14, 12))
plt.title("Crimes By District")
plt.xlabel("Frequency of Crimes")
plt.ylabel("District #")
ax = crime_by_district.plot(kind='barh')
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
```

```python
crime_by_type = no_null_crime_df['Primary Type'].value_counts(ascending=True)
fig=plt.figure(figsize=(18, 16))
plt.title("Crimes by Type")
plt.xlabel("Crime Frequency")
plt.ylabel("Type of Crime")
ax = crime_by_type.plot(kind='barh')
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
plt.savefig("CrimesByType.Png")
```
